```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.behindthename.com/namesakes/category/islamic-caliphs/chrono'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')
```

```python
specific_tables = soup.find('table', class_ = 'r1r2')
print(specific_tables)
```

```python
titles = specific_tables.find_all('th')
print(titles)
```

```python
tables_titles = [world_titles.text.strip() for world_titles in titles]
print(tables_titles)
```

```python
import pandas as pd
```

```python
dataframes = pd.DataFrame(columns = tables_titles)
print(len(dataframes))
dataframes
```

```python
column_data = specific_tables.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    world_table_titles = [data.text.strip() for data in row_data]
    if len(world_table_titles) > len(dataframes.columns):
        world_table_titles = world_table_titles[:len(dataframes.columns)]  # Trim extra data
    elif len(world_table_titles) < len(dataframes.columns):
        world_table_titles.extend([""] * (len(dataframes.columns) - len(world_table_titles)))  # Fill missing data
    
    print(world_table_titles)
    length = len(dataframes)
    dataframes.loc[length] = world_table_titles
```
